# Creating a perceptron from scratch

## Perceptron Class

```python
import numpy as np


class Perceptron:
    """Perceptron classifier

    Parameters:
    -----------
    eta: float
        Learning rate (set between 0.0 and 1.0)
    n_iter: int
        Passes over the training dataset.
    random_state: int
        Random number generator seed for random weight
        initialization.

    Attributes:
    -----------
    # not created on initialization hence the underscore
    w_: 1d-array
        Weights after fitting.
    b_: Scalar
        Bias unit after fitting.
    errors_: list
        Number of misclassifications (updates) in each epoch.

    """

    def __init__(self, eta=0.01, n_iter=50, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.random_state = random_state

    # fit method
    def fit(self, X, y):
        """Fit training data.

        Parameters:
        --------------
        X : {array-like}, shape = [n_examples, n_features]
            Training vectors, where n_examples is the number of
            examples and n_features is the number of features.
        y: {array-like}, shape = [n_examples]
            Target values.

        Return:
        --------
        self: object
        """
        # rgen is a NumPy random number generator that we seeded with a
        # user-specified random seed so that we can reproduce previous results if desired.
        rgen = np.random.RandomState(self.random_state)
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=X.shape[1])
        self.b_ = np.float_(0.0)
        self.errors_ = []

        for _ in range(self.n_iter):
            errors = 0
            for xi, target in zip(X, y):
                update = self.eta * (target - self.predict(xi))
                self.w_ += update * xi # weigth update
                self.b_ += update # bias update
                errors += int(update != 0.0) # counting the number of time update is not 0
            self.errors_.append(errors)

        return self

    # net input z= w^Tx + b
    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_) + self.b_

    def predict(self, X):
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, 0)

```

### Loading the Iris dataset

```python
import os
import pandas as pd
```

```python
s = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'

print('From URL: ', s)
```

```python
df = pd.read_csv(s, header = None, encoding = 'utf8')
df.tail()
```

### Extracting the first 100 class labels corresponding to 50 Iris-versicolor flowers and converting the class labels to integers 1 (versicolor) and 0 (setosa)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# selecting setosa and versicolor
y = df.iloc[0:100, 4].values
y = np.where(y=='Iris-setosa', 0, 1)

# extract sepal length and petal length
X = df.iloc[0:100, [0,2]].values


# plot data
plt.scatter(X[:50, 0], X[:50, 1], color = 'red', marker='o', label='Setosa')
plt.scatter(X[50:100, 0], X[50:100, 1], color = 'blue', marker='o', label='Versicolor')

plt.xlabel('Sepal length [cm]')
plt.ylabel('Petal length [cm]')
plt.legend(loc='upper left')

plt.show()
```

### Training the perceptron algorithm on the dataset along with plot of misclassification error

```python
ppn = Perceptron(eta=0.1, n_iter=10)

# calling fit
ppn.fit(X, y)

# plot missclassification error
plt.plot(range(1, len(ppn.errors_)+1), ppn.errors_, marker='o')
plt.xlabel('Epocs')
plt.ylabel('Number of updates/misclassification errors')
plt.show()
```

#### Observation:
The perceptron converged after the sixth epoch and is at a stage where it should be able to classify the training examples perfectly.

### Function to visualize the decision boundaries for 2D datasets:

```python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):
    # setup marker generatro and color map
    markers = ('o','s','^','<')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    
    lab = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    lab = lab.reshape(xx1.shape)

    plt.contourf(xx1, xx2, lab, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    # plot class examples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y==cl, 0],
                    y=X[y==cl, 1],
                    alpha = 0.8,
                    c = colors[idx],
                    marker=markers[idx],
                    label=f'Class {cl}',
                    edgecolors='black')
        
        
```

```python
plot_decision_regions(X, y, classifier=ppn)
plt.xlabel('Sepal lenght [cm]')
plt.ylabel('Petal length [cm]')
plt.legend(loc='upper left')
plt.show()
```

#### Note
* Although perceptron classified the two classes perfectly, convergence is one of the biggest challenges of the perceptron algorithm. 
* Rosenblatt showed that perceptron rule converges if the two classes can be separated by a linear hyperplane.
* If the classes are not linearly separable thne the weights will never stop updating unless a max limit of epochs is set
